Regex expression:<br>
`r"IFC(.*?)(?=\()"` <br>

```python
# Regex Test
s = "#12=IFCDIRECTION((-1.,0.));"
search_ifc = re.compile(r"IFC(.*?)(?=\()")
result = search_ifc.search(s)
result.group(0)
```

```python
#Success!
with open("../data/example_duplex.ifc") as fp:
    lines = fp.readlines()
    search_ifc = re.compile(r"IFC(.*?)(?=\()")
    ifc_types = []
    for line in lines:
        result = search_ifc.search(line)
        if result:
            print(result.group(0))
```
